### Reordening/Sorting Levels

```python
frame
```

```python
frame.swaplevel('key1','key2')
```

```python
## sorts the data using only the values in a single level.
## When swaping levels, its not uncommon to also use `sort_index` so that the result is lexicographically sorted
frame.sort_index(level=1) # level refer to row levels, here its only 2 ['key1','key2']
```

```python
## Swapping of index's and sorting of the numerical index
frame.swaplevel(0,1).sort_index(level=0)
```

# Combine
Data contained in pandas objects can be combined together in a number of ways:
* **pandas.merge**: connects rows in DataFrames based on one or more keys.
* **pandas.conca**t: concatenates or 'stacks' together objects along an axis
* **combine_first**: instance method enables splicing together oberlapping data to fill in missing values in one object with values from another

#### merge
* '**inner**' : Use only the key combinations observed in both tables (intersction)
* '**left**': Use all key combinations found in the left table
* '**right**': Use all key combinations found in the right table
* '**outer**': Use all key combinations observed in both tables together (Union)

```python
df1= pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],
                    'data2':range(3)})
df1
```

```python
df2
```

```python
## by default, 'inner' join is performed, hence key:['c','d'] from df2 is missing
pd.merge(df1, df2, on='key')
```

```python
pd.merge(df1, df2, on='key', how='outer')
```

### Indexing with Dataframe's column (different names)


```python
df3 = pd.DataFrame({'lkey':['b','b', 'a','c','a','a','b'],
                   'data1':range(7)})
df3
```

```python
df4= pd.DataFrame({'rkey':['a','b','d'],
                   'data2':range(3)})
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer')
```

### merging on Index | Hierarchically indexed data

```python
left = pd.DataFrame({'key1':['Ohio','Ohio','Ohio','Nevada','Nevada'],
                     'key2':[2000, 2001, 2002, 2001, 2002],
                     'data': np.arange(5)})
right = pd.DataFrame(np.arange(12).reshape((6,2)),
                     index=[['Nevada','Nevada','Ohio','Ohio','Ohio','Ohio'],
                            [2001, 2000, 2000, 2000, 2001, 2002]],
                     columns= ['event1','event2']
                     )
left        
```

```python
right
```

```python
pd.merge(left, right, left_on=['key1','key2'], right_index=True, indicator=True, how='outer')
```

### merging on both index



```python
left = pd.DataFrame([[1., 2.], [3.,4.], [5., 6.]],
                    index = ['a','c','e'],
                    columns = ['ohio','Nevada'])

right = pd.DataFrame([[7., 8.], [9., 10.], [11, 12.], [13, 14]],
                     index = ['b','c','d','e'],
                     columns = ['Missouri','Alabama'])
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, left_index=True, right_index =True, how='outer',indicator=True )
```

#### Alternative: join
- Dataframe has a convenient join instance for merging by index
- It can be used to combine together many DataFrame objects having the same or similar indexes but non-overlapping columns.

```python
left.join(right, how='outer')
```


## Concatenation

```python
s1 = pd.Series([0,1], index=['a','b'])
s2 = pd.Series([2,3,4], index=['c','d','e'])
s3 = pd.Series([5,6], index = ['f','g'])
```

```python
## in the axis:1 case, since there is no columns name and there is single column on all series, it concetenate 
## them. Hence in the end there are three columns.
pd.concat([s1, s2, s3], axis=1) #axis 0/1 : row/column
```

```python
pd.concat([s1, s2, s3], axis=0)
```

```python
s1
```

## Reshaping/Pivoting
* There are number of basic operations for rearranging tabular data, alternately refered to as reshape or pivot operations.
* There are two primarily actions:
  1. stack
    - This 'rotate'or pivots from the column in the data to the rows.
  2. unstack
     - this pivots from the rows into the columns

```python
data = pd.DataFrame(np.arange(6).reshape((2,3)),
                    index = pd.Index(['Ohio','Colorado'], name='state'),
                    columns = pd.Index(['one','two','three'],
                    name='number'))
data
```

```python
result=data.stack()
result
```

```python
result.unstack()
```

```python
result.unstack(0)
```

```python
result.unstack('number')
```
