## Hierarchical Indexing
* it enable you to have multiple index levels on an axis.
* Somewhat abstractly, it provides a way for you to work with higher dimensionanal data in a lower dimensional form.

```python
import pandas as pd
import numpy as np
```

## Hierachical indexing in *dataframe*

```python
frame = pd.DataFrame(np.arange(12).reshape(4,3),
                     index=[['a','a','b','b'],[1,2,1,2]],
                     columns=[['Ohio','Ohio','Colorado'], 
                              ['Green','Red','Green']])
frame
```

```python
frame.index.names = ['key1','key2']
frame
```

```python
frame.columns.names = ['state','color']
frame
```

```python
  frame['Ohio']
```

### Reordening/Sorting Levels

```python
frame
```

```python
frame.swaplevel('key1','key2')
```

```python
## sorts the data using only the values in a single level.
## When swaping levels, its not uncommon to also use `sort_index` so that the result is lexicographically sorted
frame.sort_index(level=1) # level refer to row levels, here its only 2 ['key1','key2']
```

```python
## Swapping of index's and sorting of the numerical index
frame.swaplevel(0,1).sort_index(level=0)
```

### Summary statistics by Level
* we can aggregate on level by a particular axis 
* We can aggregate by level on either the rows or columns



```python
frame
```

```python
frame.sum(level='key1') #key2
```

```python
frame.sum(level='key2', axis=0)
```

```python
frame.sum(level='color', axis=1)
```
