```python
import pandas as pd
import numpy as np
```

## Upload dataset

```python
data_url = 'http://bit.ly/2cLzoxH'
gapminder = pd.read_csv(data_url)
print(gapminder.head(3))
```

## Mean life expectency per continent

```python
df = gapminder[['continent','lifeExp']]
df.head(2)
```

```python
pd.pivot_table(df, values='lifeExp',columns='continent')
```

```python
pd.pivot_table(df, values='lifeExp',columns='continent')
```

```python
pd.pivot_table(df, values='lifeExp',columns='continent').stack().to_frame().reset_index()
```

## Mean lifeexp per continent per year

```python
df = gapminder[['continent','year','lifeExp']]
df.head(2)
```

```python
pd.pivot_table(df, values='lifeExp',index=['year'], columns='continent').reset_index()
```

## Minimum Life Expectancy per continent per year

```python
pd.pivot_table(df, values='lifeExp', index=['year'], columns='continent',aggfunc='min')
```

## Minimun, maximum Life Expectancy per Continent per Year

```python
pd.pivot_table(df, values='lifeExp', index=['year'],columns='continent', aggfunc=['min','max'])
```
